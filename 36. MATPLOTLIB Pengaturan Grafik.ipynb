```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,20,0.5)
y = x**2
```

```python
fig = plt.figure(figsize=(4,4), dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y);
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi=200)
axes[0].plot(x,y)
axes[0].set_title('grafik 1')

axes[1].plot(x,y)
axes[1].set_title('grafik 2')

plt.tight_layout()

plt.show()
```

```python
# menyimpan gambar yang ingin dicetak

fig.savefig('gambarku.png', dpi=300)
```

```python
# menambahkan legenda (keterangan dalam sebuah grafik)

fig = plt.figure(dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='pangkat 2')
ax.plot(x,x**3, label = 'pangkat 3')

#menambahkan legend
ax.legend(loc=(0.2,0.2)) #20% kiri 20% bawah
plt.show()
```

```python
fig = plt.figure(facecolor='yellow', dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='pangkat 2')
ax.plot(x,x**3, label = 'pangkat 3')
ax.plot(x,x**4, label = 'pangkat 4')

ax.legend(loc=(0.2,0.2)) #20% kiri 20% bawah
ax.grid(True) #menambahkan grid
plt.show()
```

```python
# mengatur garis ketebalan dll

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-') #linewidth ketebalan garis
ax.plot(x,x**3, label = 'pangkat 3', lw=2, ls='--')
ax.plot(x,x**4, label = 'pangkat 4', lw=3.5, ls='-.')

ax.legend(loc=(0.2,0.2)) #20% kiri 20% bawah
ax.grid(True) #menambahkan grid

plt.show()
```

```python
# mengatur marker
# markerfacecolor = mfc
# markeredgecolor = mec
# markeredgewidth = mew

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-', marker='o', markersize=5, mfc='yellow', mew=1, mec='red') #marker
ax.plot(x,x**3, label = 'pangkat 3', lw=2, ls='--')
ax.plot(x,x**4, label = 'pangkat 4', lw=3.5, ls='-.')

ax.legend(loc=(0.2,0.2)) #20% kiri 20% bawah
ax.grid(True) #menambahkan grid

plt.show()
```

```python
# membatasi x dan y

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label='pangkat 2', lw=1, ls='-', marker='o', markersize=5, mfc='yellow', mew=1, mec='red') #marker
ax.plot(x,x**3, label = 'pangkat 3', lw=2, ls='--')
ax.plot(x,x**4, label = 'pangkat 4', lw=3.5, ls='-.')

ax.legend(loc=0) #20% kiri 20% bawah
ax.grid(True) #menambahkan grid

ax.set_xlim([0,12.5]) #membatasi x
ax.set_ylim([0,20000]) #mebatasi y

plt.show()
```
