# loc
## loc['nama baris', 'nama kolom']

jika kita tidak tahu nomor indeks atau tidak berniat memakai indeks

```python
data_baru = pd.read_csv('kapal_titanic.csv', index_col='embarked')
data_baru
```

```python
data_baru.loc['S'] #loc tidak harus berupa list
```

```python
data_baru.loc['S', 'age']
```

```python
data_baru.loc['S', ['age', 'fare']] #jika tidak mau colom age saja ingin kolom lainnya maka menggunakan list
```

```python
data_baru.loc[['S', 'Q'], ['age', 'fare']]
```
