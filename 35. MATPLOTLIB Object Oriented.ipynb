```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot # adalah cara yang sangat konvensional
# Object Oriented adalah cara yang poerfull
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
x
```

```python
y
```

# Versi matplotlib OOP

```python
# versi 1

fig = plt.figure()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 0.1 artinya 10% dari kiri #axes adalah object oriented
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y) #sampai disini bisa di print
axes.set_xlabel('sumbu x')
axes.set_ylabel('sumbu y')
axes.set_title('judul grafik')
plt.show()
```

```python
# membuat axes di dalam axes

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])

axes1.plot(x,y, color='red')
axes1.set_xlabel('sumbu X besar')
axes1.set_ylabel('sumbu Y besar')
axes1.set_title('judul grafik pertama')

axes2.plot(y,x, color='blue')
axes2.set_xlabel('sumbu X kecil')
axes2.set_ylabel('sumbu Y kecil')
axes2.set_title('judul grafik kecil')

plt.show()
```

```python
# versi 2

fig, axes = plt.subplots(3,2) # 3 baris dan 2 kolom
fig.tight_layout() #agar jarak tidak berdempetan
```

```python
fig, axes = plt.subplots(1,2) 

# Plot dor the first axes
axes[0].plot(x,y)
axes[0].set_xlabel('sumbu X1')
axes[0].set_ylabel('sumbu Y1')
axes[0].set_title('judul satu')

# Plot dor the second axes
axes[1].plot(y,x)
axes[1].set_xlabel('sumbu X2')
axes[1].set_ylabel('sumbu Y2')
axes[1].set_title('judul dua')

fig.tight_layout()
```
