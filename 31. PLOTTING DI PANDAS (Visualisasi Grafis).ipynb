```python
import numpy as np
```

```python
np.random.seed(100)
```

```python
angka = []
for i in range (0,100):
    angka.append(i)
```

```python
baru = pd.DataFrame(np.random.rand(100,5), angka, ['A', 'B', 'C', 'D', 'E'])
```

```python
baru
```

```python
baru.plot.area() #kolom area luas permukaan
```

```python
baru.plot.area(alpha=0.4)
```

```python
baru2 = pd.DataFrame(np.random.rand(4,5), [0,1,2,3], ['A', 'B', 'C', 'D', 'E'])
baru2
```

```python
baru2.plot.bar(alpha=0.4)
```

```python
baru2.plot.bar(stacked=True)
```

```python
baru2 ['C'].plot.bar()
```

```python
baru.plot.scatter(x='A', y='B')
```

```python
baru.plot.scatter(x='A', y='B', c='E') # c disini adalah warna yang sesuai kolom E
```

```python
baru.plot.scatter(x='A', y='B', c='E', cmap='ocean') #cmap membedakan berdasarkan warna
```

```python
# membedakan berdasarkan ukuran

baru.plot.scatter(x='A', y='B', s=baru.D)
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D*100)
```

```python
baru.plot.scatter(x='A', y='B', s=baru.D*100, c='C', cmap='Reds')
```

```python
baru.plot.box() #melihat persentil
```

```python
baru.plot.line(y='B')
```

```python
baru.plot.line(y='B', lw=0.5 ) #menentukan tebal tipisnya garis
```

```python
baru.plot.line(y='B', lw=0.5, figsize=(10,5) ) #panjang lebarnya garis
```

```python
baru.plot.hexbin(x='A', y='B') #berbentuk heksagonal
```

```python
baru.plot.hexbin(x='A', y='B', gridsize=20)
```
