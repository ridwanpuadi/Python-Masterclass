```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
dataku.head()
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# heatmap (matrix plot) = data yang bersifat numerik
# tabel harus sudah berupa matrix
# indeksnya harus sudah berupa kategori melalui corr()

dataku.corr()
```

```python
# kelemahannya adalah di data ini kita hanya bisa mengira-ngira

sns.heatmap(dataku.corr());
```

```python
sns.heatmap(dataku.corr(), annot=True);
```

```python
sns.heatmap(dataku.corr(), annot=True, cmap='Greens');
```

```python
dataku.head()
```

```python
# jika kita tidak ingin menggunakan corr() untuk menghilangkan indeks maka bisa dengan cara

dataku2 = pd.read_csv('kapal_titanic.csv', index_col='embarked')
dataku2
```

```python
sns.heatmap(dataku2.iloc[:,[0,1,4]].head(), annot=True, cmap='Greens');
```

```python
# heatmap advanced

terbang = sns.load_dataset('flights')
terbang.head()
```

```python
terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
terbang_pivot = terbang.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(terbang_pivot, cmap='Blues')
```

```python
sns.heatmap(terbang_pivot, cmap='Blues', linecolor='black', linewidths=0.5);
```

```python
# Diagram pohon clustering disebut dengan Dendogram

sns.clustermap(terbang_pivot)
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm');
```

```python
sns.clustermap(terbang_pivot, cmap='coolwarm', standard_scale=1);
```
