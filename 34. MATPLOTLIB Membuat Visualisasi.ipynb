```python
import numpy as np

x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y)
plt.xlabel('sumbu x')
plt.ylabel ('sumbu y')
plt.title ('grafik')
```

```python
# membuat dua grafik berdampingan

plt.subplot(121)
plt.plot(x,y, color='red')
plt.xlabel('sumbu x1')
plt.ylabel ('sumbu y1')

plt.subplot(122)
plt.plot(y,x, color='blue')
plt.xlabel('sumbu x2')
plt.ylabel ('sumbu y2')
plt.tight_layout() #untuk spasi agar tidak terlalu berdempetan
plt.show()
```
