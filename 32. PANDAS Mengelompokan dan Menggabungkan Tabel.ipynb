```python
# menggabungkan beberapa baris/kolom berdasarkan kesamaan

bisnis = pd.DataFrame({'Perusahaan':['Telkom', 'Telkom', 'Pertamina', 'BRI', 'Pertamina', 'KFC', 'KFC', 'Indosat'],
                          'Karyawan':['Joko', 'Budi', 'Susi', 'Iriana', 'Bambang', 'Burhanudin', 'Intan', 'Mukiman'],
                          'Usia':[44,53,67,34,60,19,20,17]})
bisnis
```

```python
bisnis.groupby('Perusahaan')
```

```python
bisnis.groupby('Perusahaan').sum() #fungsi khusus untuk numerik maka yang akan muncul hanya usia saja
```

```python
# misal telkom yaitu 53+44 = 97, KFC yaitu 19+20
```

```python
bisnis.groupby('Perusahaan').count() # ingin mengetahui jumlah datanya
```

```python
bisnis.groupby('Perusahaan').mean() 
```

```python
bisnis.groupby('Perusahaan').min() # muncul berdasarkan alfabet
```

```python
bisnis.groupby('Perusahaan').max() 
```

```python
bisnis.groupby('Perusahaan').Usia.min()
```

```python
bisnis.groupby('Perusahaan').describe()
```

```python
bisnis.groupby('Perusahaan').describe().transpose()
```

```python
# frame bentuknya sama seperti SQL

satu = pd.DataFrame ({'key':['k0', 'k1', 'k2', 'k3', 'k4'],
                         'one': [1,2,3,4,5],
                         'two': [6,7,8,9,10]})
```

```python
dua = pd.DataFrame ({'key':['k0', 'k1', 'k2', 'k3', 'k4'],
                         'three': [1,2,3,4,5],
                         'four': [6,7,8,9,10]})
```

```python
satu
```

```python
dua
```

```python
# menggunakan dua parameter berdasarkan key, key disini harus unique
# menggunakan merge

baru = pd.merge(satu, dua, how='inner', on='key')
baru
```

```python
satu = pd.DataFrame ({'key1':['K0', 'K0', 'K1', 'K2'],
                      'key2':['K0', 'K1', 'K0', 'K1'],
                          'A':['A0', 'A1', 'A2', 'A3'],
                          'B':['B0', 'B1', 'B2', 'B3']})

dua = pd.DataFrame ({'key1':['k0', 'k1', 'k1', 'k2'],
                     'key2':['k0', 'k0', 'k0', 'k0'],
                          'C':['C0', 'C1', 'C2', 'C3'],
                          'D':['D0', 'D1', 'D2', 'D3']})
```

```python
satu
```

```python
dua
```

```python
baru2 = pd.merge(satu, dua, how='left', on=['key1', 'key2'])
baru2
```

```python
baru2 = pd.merge(satu, dua, how='right', on=['key1', 'key2'])
baru2
```

```python
baru2 = pd.merge(satu, dua, how='outer', on=['key1', 'key2'])
baru2
```

```python
# keys nya berupa indeks nya
# dengan menggunakan join

bapak = pd.DataFrame ({'A':['A0', 'A1', 'A2'],
                      'B':['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

ibu = pd.DataFrame ({'C':['C0', 'C1', 'C2'],
                      'D':['D0', 'D1', 'D2']},
                      index=['K0', 'K2', 'K3'])
```

```python
bapak
```

```python
ibu
```

```python
bapak.join(ibu)
```

```python
bapak.join(ibu, how='inner')
```

```python
bapak.join(ibu, how='left')
```

```python
bapak.join(ibu, how='right')
```

```python
bapak.join(ibu, how='outer')
```
