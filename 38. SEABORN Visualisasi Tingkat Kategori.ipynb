```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('kapal_titanic.csv')
dataku.head()
```

```python
# untuk memunculkan data kategori saja

dataku.describe(include='O')
```

# Barplot (histogram)

```python
sns.barplot(x='sex', y='fare', data=dataku);
```

```python
sns.barplot(x='fare', y='sex', data=dataku);
```

```python
# merubah estimator

sns.barplot(x='sex', y='fare', data=dataku, estimator=sum);
```

```python
sns.barplot(x='sex', y='fare', data=dataku, estimator=np.std);
```

```python
# countplot

sns.countplot(x='sex', data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', data=dataku);
```

```python
sns.countplot(x='deck', hue='sex', order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], data=dataku);
```

```python
# Boxplot (box & whiskers plot)
# yang diatas dan dibawah garis adalah nilai outlier
# outlier atas = Q3+1.5*(Q3-Q1)
# Outlier bawah = Q1-1.5*(Q3-Q1)
# Q3-Q1 = Interquartile range (IR) batas masing2 quartal

sns.boxplot(x='sex', y='fare', data=dataku);
```

```python
sns.boxplot(x='sex', y='fare', hue='deck', hue_order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], data=dataku);
```

# Violin Plot

```python
sns.violinplot(x='sex', y='fare', data=dataku);
```

```python
sns.violinplot(x='sex', y='fare', hue='deck', hue_order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], data=dataku);
```

```python
sns.violinplot(x='deck', y='fare', hue='sex', order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], split=True, data=dataku);

#split bisa digunakan ketika variabel lebih dari 2 kategori dalam hal ini adalah sex
```

# strip plot

```python
sns.stripplot(x='deck', y='fare', hue='sex', order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], jitter=True, data=dataku);

# untuk data bisa lebih nyebar pakai 'jitter'
```

# Swarmplot

```python
sns.swarmplot(x='deck', y='fare', hue='sex', order=['A', 'B', 'C', 'D', 'E', 'F', 'G'], data=dataku);
```
